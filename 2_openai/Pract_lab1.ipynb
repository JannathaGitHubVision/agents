```python
from openai import OpenAI
import json
import math
```

```python
client = OpenAI()

def calculate_square_root(x: float) -> float:
    return math.sqrt(x)

# 2) Define the tool schema (how the model sees the function)
tools = [
    {
        "type": "function",
        "name": "calculate_square_root",
        "function": {
            "name": "calculate_square_root",
            "description": "Calculate the square root of a number.",
            "parameters": {
                "type": "object",
                "properties": {
                    "x": {
                        "type": "number",
                        "description": "The number to take the square root of."
                    }
                },
                "required": ["x"]
            },
        },
    }
]


user_message = "What is the square root of 16?"

# 3) Call the model, allowing it to use the tool
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": user_message}],
    tools=tools
)


message = response.choices[0].message

if message.tool_calls:
    tool_call = message.tool_calls[0]
    print(tool_call)
    tool_name = tool_call.function.name
    tool_args = json.loads(tool_call.function.arguments)

    if tool_name == "calculate_square_root":
        result = calculate_square_root(**tool_args)
        print(f"Tool Result : {result}")
else:
    print(message.content)
```
